4. Tenemos una opción para el 5 % de tasa de interés:

Podemos usar el símbolo de negación (~) con nuestro filtro actual para darnos todo EXCEPTO lo que hemos definido en ese filtro. Esto funciona porque solo tenemos dos tasas de interés (cualquier valor que NO es 3 % es 5 %).

```python
filtro3=dt['Interest Rate'] == 0.05
dt_3=dt[filtro1 & filtro3]
dt_3
```

5. Puesto que queremos graficar la suma acumulativa del interés pagado, podemos usar la función .cumsum().   ¿Cuáles son los tres primeros valores en la salida para la suma acumulativa de la tasa del 3 %?

```python
acum=dt_2['Interest Paid'].cumsum()

acum[0:3]
```

```python
acum2=dt_3['Interest Paid'].cumsum()
```

```python
plt.figure()
plt.plot(dt_3['Month'], acum2, c='k', label='interés pagado 5%')
plt.plot(dt_2['Month'], acum, c='b', label='interés pagado 3%')
plt.legend()

plt.show()
```
